```python
import pandas as pd
```

```python
dataset_path = "facts_countries.csv"
df = pd.read_csv(dataset_path, sep=";", skiprows=[1])
df
```

```python
for column in df.columns:
    df[column].fillna(0, inplace=True)

df
```

```python
df_transpose = df.transpose()

df_transpose
```

```python
df_transpose.columns = df_transpose.iloc[0]
df_transpose
```

```python
df_transpose = df_transpose[1:]
df_transpose
```

```python
df_transpose.columns = map(str.lower, df_transpose.columns)
df_transpose
```

```python
df_transpose.columns = map(
    lambda column: column.replace(" ", "_"), df_transpose.columns
)
df_transpose
```
